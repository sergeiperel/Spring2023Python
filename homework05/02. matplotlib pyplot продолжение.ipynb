# Заголовок

Это формула гауссового распределения: $e^{-x^2/2}$  
$\alpha$, $\beta$, $\pi = 3.14$

```python
def axis_settings(ax, title):
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(title)

x = np.linspace(-5, 5, 5)
y = np.ones_like(x)

fig, axes = plt.subplots(1, 4, figsize=(16,4))

#line widths
linewidths = [0.5, 1.0, 2.0, 4.0]
for n, linewidth in enumerate(linewidths):
    axes[0].plot(x, y + n, color='blue', linewidth=linewidth)
axis_settings(axes[0], 'line width')

#Line style
linestyles = ['-', '--', '-.', ':']
for n, linestyle in enumerate(linestyles):
    axes[1].plot(x, y + n, color='blue', linestyle=linestyle)
axis_settings(axes[1], 'line style')

#Marker type
markertypes = ['+', '*','o','.','s','1','2','3','4']
for n, marker in enumerate(markertypes):
    axes[2].plot(x, y + n, color='blue', marker=marker, ls='')
axis_settings(axes[2], 'marker type')

#Marker size and color
markertypes = [(4, "white"), (8, "red"), (12, "yellow"), (16, "lightgreen")]
for n, markertype in enumerate(markertypes):
    markersize = markertype[0]
    markercolor = markertype[1]
    axes[3].plot(x, y + n, color='blue', ls='', marker='o', 
                 markersize=markersize, markerfacecolor=markercolor, markeredgecolor='black')
axis_settings(axes[3], 'marker type')
```

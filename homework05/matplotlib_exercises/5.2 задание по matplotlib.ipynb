## Задание 1  

1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!

```python
import matplotlib.pyplot as plt
import numpy as np
def func(x):
    return x**2 * np.sin(1/x**2) + x
ox = np.linspace(-1, 1, 250) 

plt.plot(ox, func(ox))
plt.title('$y = x^2⋅sin(1/x^2)+x$')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)

```python
def func_semilogy(x):
    return abs(1 - (x ** 2)/(1 + x ** 2))
ox2 = np.linspace(5, 25, 250) 

plt.figure(figsize = (10, 5))
plt.semilogy(ox2, func_semilogy(ox2))
plt.title('$y = |1 - (x^2)/(1 + x^2)|$')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>! Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.

```python
from scipy.special import j1, i0e, i1e

# Bessel function of the first kind of order 1.


x = np.linspace(-10, 10, 1000)
plt.figure(figsize = (7, 7))
plt.title('Faster versions of common Bessel functions')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, j1(x), color = 'red', label = "Bessel function of the first kind of order 1")
plt.plot(x, i0e(x), color = 'green', label = "Exponentially scaled modified Bessel function of order 0")
plt.plot(x, i1e(x), color = 'blue', label = "Exponentially scaled modified Bessel function of order 1")

plt.grid()
plt.legend(loc='best')
plt.show()



```

```python
from scipy.special import spherical_jn
x = np.arange(0.0, 10.0, 0.01)
fig, ax = plt.subplots()
ax.set_ylim(-0.5, 1.5)
ax.set_title(r'Spherical Bessel functions $j_n$')
for n in np.arange(0, 4):
    ax.plot(x, spherical_jn(n, x), label=rf'$j_{n}$')
plt.legend(loc='best')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

## Exercise 2  

1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
def func_sin(x):
    return np.sin(x) / x

ox = np.linspace(-6*np.pi, 6*np.pi, 250)

plt.figure(figsize=(15, 8))
plt.plot(ox, func_sin(ox))

x_axis = [(str(-12+i) + 'π/2' if i != 12 else '0') for i in range(25)]
plt.xticks(np.arange(-6*np.pi, 6*np.pi+np.pi/2, np.pi/2), x_axis)
plt.xlabel('x-axis (rad))')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
import matplotlib as mpl

x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.sin(Y)

fig, ax = plt.subplots(figsize=(6,5))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = ax.pcolor(X, Y, Z, norm = norm, cmap='inferno')
ax.axis('tight')

plt.title('Цветовой график $sin(x)sin(y)$')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
cb = fig.colorbar(p, ax=ax)
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-X**2 - Y**2)
fig, ax = plt.subplots(figsize=(6,5))
plt.imshow(Z, cmap='rainbow', extent=[-3, 3, -3, 3])
plt.colorbar()
plt.title('Цветовой график $e^{-x^2-y^2}$')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

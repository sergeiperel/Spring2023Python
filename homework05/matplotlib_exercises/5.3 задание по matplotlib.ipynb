## Задание 1
Полиномиальная регрессия 

```python
import numpy as np
import numpy.linalg
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

def approx(n):
    dots = np.linspace(1, 15, n)

    a = np.array([[x**j for j in range(n)] for x in dots])        
    b = f(dots)

    w = np.linalg.solve(a, b)  # точки в полиноме 
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  # решение
    
    return res


x = np.linspace(1, 15, 250)
plt.grid()
plt.title('Исходная функция')
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('y')
plt.show()

n = 15
plt.grid()
plt.title(f'Точки, по которым строился многочлен при n = {n}')
plt.plot(np.linspace(1, 15, n), f(np.linspace(1, 15, n)), '8')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.grid()
plt.title(f'Полиномиальная аппроксимация при n = {n}')
plt.plot(np.linspace(1, 15, n), approx(n))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Мы видим, что при увеличении числа n функция полиномиальной аппроксимации становится более гладкой и похожей на оригинальную функцию, тогда как при уменьшении числа n она становится ломаной и непохожей на оригинальную функцию.

## Задание 2
Полиномиальная регрессия с зашумлённым данными 

```python
import numpy.random

def approx(n):
    dots = np.linspace(1, 15, n)

    a = np.array([[x**j for j in range(n)] for x in dots])        
    b = f(dots) + np.random.normal(0, 0.01)

    w = np.linalg.solve(a, b)  # точки в полиноме 
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  # решение
    
    return res


x = np.linspace(1, 15, 250)
plt.grid()
plt.title('Исходная функция')
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('y')
plt.show()

n = 15
plt.grid()
plt.title(f'Точки, по которым строился многочлен при n = {n}')
plt.plot(np.linspace(1, 15, n), f(np.linspace(1, 15, n)), '8')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.grid()
plt.title(f'Полиномиальная аппроксимация с шумом при n = {n}')
plt.plot(np.linspace(1, 15, n), approx(n))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Задание 3
Метод наименьших квадратов

```python
N = 100
n = 3
dots = np.linspace(1, 15, N)
a = np.array([[x**j for j in range(n)] for x in dots])
b = f(dots) + np.random.normal(0, 0.1)

w, resid, rank, s = numpy.linalg.lstsq(a, b)  # точки в полиноме
res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  # решение

x = np.linspace(1, 15, 250)
plt.grid()
plt.title('Исходная функция')
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.grid()
plt.title(f'Точки, по которым строился многочлен при n = {n}')
plt.plot(dots, f(dots), '8')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.grid()
plt.title(f'Метод наименьших квадратов при n = {n}')
plt.plot(dots, res)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Как можно увидеть, метод наименьших квадратов даёт не точный график при n=3. При увеличении n график становится более точным.

Из этой работы можно увидеть, что при достаточно большой степени полинома n график можно успешно аппроксимировать (даже с помехами).
